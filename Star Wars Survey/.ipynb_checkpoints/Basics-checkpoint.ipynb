# Star Wars Survey

This project is focused on cleaning and exploring the data set that was collected by the team at [FiveThirtyEight](https://fivethirtyeight.com/) regarding [Star Wars: The Force Awakens](https://en.wikipedia.org/wiki/Star_Wars:_The_Force_Awakens). The data can be found [here](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey).

We'll be focused on extracting insights after cleaning and exploring the data set.

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")

star_wars.head(10)
```

```python
star_wars.columns
```

## Cleaning the Data

We will be removing any null RespondantID values since they may be innaccurate data if it was electronically collected. Before that, we will save the data from the first row which contains information about the unnamed columns.

```python
view_names = star_wars.iloc[0,3:9]
ranking_names = star_wars.iloc[0,9:15]
preference_names = star_wars.iloc[0,15:29]
```

```python
# Drop first row which has a null RespondentID
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
```

Next, we will be mapping the Yes/No Columns to boolean values.

```python
yes_no = {
    "Yes": True,
    "No": False
}

# Map values
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = \
    star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)

star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = \
    star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

Third, we will be mapping the checkbox columns to boolean values.

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.nan: False
}

# Map values
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
    
view_rename = dict(zip(star_wars.columns[3:9], view_names + '_viewed'))    
star_wars = star_wars.rename(columns=view_rename)
```

Fourth, we will be renaming the ranking columns.

```python
# Turn into numerical type
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

ranking_rename = dict(zip(star_wars.columns[9:15], ranking_names + '_ranked'))
star_wars = star_wars.rename(columns=ranking_rename)
```

Finally, we will be changing the preference column into numerical values.

```python
numeric_ranking_map = {
    'Very favorably': 5,
    'Somewhat favorably': 4,
    'Neither favorably nor unfavorably (neutral)': 3,
    'Somewhat unfavorably': 2,
    'Very unfavorably': 1
}

# Map values
for col in star_wars.columns[15:29]:
    star_wars[col] = star_wars[col].map(numeric_ranking_map)

preference_rename = dict(zip(star_wars.columns[15:29], preference_names))

star_wars = star_wars.rename(columns=preference_rename)
```

## Analyzing and Plotting the Data

Now that the data is cleaned, we can begin analysis.

### Finding the Highest-Ranked Movie

Let's see if we can find the highest-ranked movie.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib

matplotlib.style.use('fivethirtyeight')

plt.bar(ranking_names, star_wars[star_wars.columns[9:15]].mean())
plt.xticks(rotation=90)
plt.ylabel('Average Ranking Preference')
```

It seems that the higher ranked movies are the last 3 episodes (Star Wars Episodes 4-6). The values above are based off the ranking preference (so the lower the value the better).

### Finding the Most-Viewed Movie

```python
plt.bar(view_names, star_wars[star_wars.columns[3:9]].sum())
plt.xticks(rotation=90)
plt.ylabel('Views')
```

As seen above, it seems that the higher-ranked movies (Star Wars Episodes 4-6) are also the most viewed.

### Splitting by Gender

First set of tables will be rankings, and second set of tables will be viewings.

```python
males = star_wars[star_wars['Gender'] == 'Male']
females = star_wars[star_wars['Gender'] == 'Female']

plt.figure(figsize=(11,8))

plt.subplot(2, 2, 1)
plt.bar(range(6), males[males.columns[9:15]].mean())
plt.tick_params(labelbottom=False)   
plt.title('Male Rankings')
plt.ylabel('Average Ranking Preference')

plt.subplot(2, 2, 2)
plt.bar(range(6), females[females.columns[9:15]].mean())
plt.tick_params(labelbottom=False)   
plt.title('Female Rankings')
plt.ylabel('Average Ranking Preference')

plt.subplot(2, 2, 3)
plt.bar(view_names, males[males.columns[3:9]].sum())
plt.xticks(rotation=90)
plt.ylabel('Views')

plt.subplot(2, 2, 4)
plt.bar(view_names, females[females.columns[3:9]].sum())
plt.xticks(rotation=90)
plt.ylabel('Views')
plt.show()
```

While more males watched episodes 1-3, males liked them less than females did.

### Analyzing Education, Location, and Perception

In this part, we will be looking at the perception of which character shot first depending on education and location. First, we will drop rows where survey respondents have not seen Star Wars. Previously, we did not drop them since their presence did not affect the average or counts.

We will also look at the counts.

```python
star_wars_seen = star_wars[star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] == True]
```

```python
star_wars_seen['Location (Census Region)'].value_counts()
```

```python
# We will drop East South Central since it is less than 5% of the values.
star_wars_seen = star_wars_seen[star_wars_seen['Location (Census Region)'] != 'East South Central']
```

```python
star_wars_seen['Education'].value_counts()
```

```python
# We will drop the Less than high school degree
star_wars_seen = star_wars_seen[star_wars_seen['Education'] != 'Less than high school degree']
```

```python
star_wars_seen['Which character shot first?'].value_counts()
```

Now we'll calculate the crosstab percentages per row below.

```python
columns = [
    'Location (Census Region)',
    'Education',
    'Which character shot first?'
]

crosstab = pd.crosstab(
    index = [star_wars_seen['Location (Census Region)'],star_wars_seen['Education']],
    columns = star_wars_seen['Which character shot first?'],
    normalize='index'
).applymap(lambda x: round(x*100,0))
crosstab
```

```python
crosstab[:17].plot(kind='bar',figsize=(11,5)).legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
crosstab[17:33].plot(kind='bar',figsize=(11,5)).legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

Visually, this seems to be a lot to take in. However, we can see a lot of interesting trends in certain regions such as an overwhelming answer for one character (Han or Greedo) or even confusion (not understanding the question).

We will address the values by looking only by education or location.

```python
crosstab_location = pd.crosstab(
    index = star_wars_seen['Location (Census Region)'],
    columns = star_wars_seen['Which character shot first?'],
    normalize='index'
).applymap(lambda x: round(x*100,0))
crosstab_location.plot(kind='bar',figsize=(11,5), ylabel='Percent').legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
crosstab_education = pd.crosstab(
    index = star_wars_seen['Education'],
    columns = star_wars_seen['Which character shot first?'],
    normalize='index'
).applymap(lambda x: round(x*100,0))
crosstab_education.plot(kind='bar',figsize=(11,5), ylabel='Percent').legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

From the above plots, we can see that most locations believe that Han shot first. As for education, most education levels believe that Han also shot first.

### Character Popularity

Finally, we'll answer the following questions:
- Which character do respondents like the most?
- Which character do respondents dislike the most?
- Which character is the most controversial (split between likes and dislikes)?

```python
star_wars[preference_names].describe()
```

```python
# Using the dictionary's values in order to select the preference columns
characters = star_wars[preference_rename.values()].mean().sort_values(ascending=False)
characters
```

```python
star_wars[preference_names].plot(kind='box',rot=90)
```

From the above, we see that the most liked character is Han Solo while the least liked is Jar Jar Binks. The most controversial is Darth Vader due to the highest standard deviation.

## Conclusion

In conclusion, we found the following:
- The highest ranked and most viewed movies were the last 3 episodes (4-6).
- More males watched episodes 1-3 but liked them less than females did.
- Most location and education gruops believe that Han shot first, but there were a substantial number that didn't understand the question.
- The most liked character was Han Solo, and the least liked was Jar Jar Binks.
- The most controversial character was Darth Vader.
