## Cleaning Column Names

We should convert the column names to snakecase. The below section shows that process starting with the array of column names.

```python
print(autos.columns)
```

```python
autos.rename(columns={'yearOfRegistration':'registration_year', 
                      'monthOfRegistration':'registration_month', 
                      'notRepairedDamage':'unrepaired_damage',
                      'dateCreated':'ad_created'
                     },
             inplace=True
            )

import re

# Substitute capital letter with an underscore and the lower case of that letter
for column in autos.columns:
    autos.rename(columns={column:re.sub('([A-Z]+)', r'_\1',column).lower()},inplace=True)

autos.head()
```

Certain columns were re-named directly and others were converted to snake_case using regex. This makes the column names into the preferred format for Python.

## Initial Exploration and Cleaning

```python
autos.describe(include='all')
```

Looking at the dataframe, you can that columns with a single value are candidates to be dropped (e.g. nr_of_pictures).

Some columns that are stored as text but are numeric need to be cleaned (e.g. price and odometer).

Also, columns with 2 unique values but where 1 of those values has a frequency of almost near the count need to be investigated (e.g. seller and offer_type).

### Cleaning the Price and Odometer Columns

```python
autos['price'] = autos['price'].str.replace('$','',regex=False).str.replace(',','',regex=False).astype(int)

autos.describe()
```

```python
autos['odometer'] = autos['odometer'].str.replace(',','').str.replace('km','').astype(int)

autos.rename(columns={'odometer':'odometer_km'},inplace=True)
```

```python
autos.describe()
```

Now, we will look at outliers and remove them if necessary in the odometer_km and price columns.

```python
print(autos['price'].unique().shape)
print(autos['odometer_km'].unique().shape)
```

```python
print(autos['price'].value_counts().sort_index(ascending=True).head())
print(autos['price'].value_counts().sort_index(ascending=False).head())
print(autos['odometer_km'].value_counts())
```

We can see from the 3 outputs that some values need to be removed. For price, it is unrealistic to have a price of 0 or a price too high. For odometer values, we will just remove any NaN values.

We will use the 25th and 75th percentiles that were found in the earlier data frame describe() for the price.

```python
autos = autos[autos["price"].between(1100,7200)]
autos = autos.dropna(subset=['odometer_km'])
```

```python
autos.describe()
```

### Exploring the Date Columns

We will be looking at the columns: date_crawled, ad_created, and last_seen

```python
autos['date_crawled'] = autos['date_crawled'].str[:10]
autos['ad_created'] = autos['ad_created'].str[:10]
autos['last_seen'] = autos['last_seen'].str[:10]
```

```python
print(autos['date_crawled'].value_counts(normalize=True, dropna=False).sort_index())
print(autos['ad_created'].value_counts(normalize=True, dropna=False).sort_index())
print(autos['last_seen'].value_counts(normalize=True, dropna=False).sort_index())
```

Most of the values seem pretty uniform except the last_seen which is skewed towards the later dates.

```python
autos['registration_year'].describe()
```

From the above, we can see a slight skew to the left in the data distribution. Values after 2016 cannot be registered after the listing was seen and values before 1900 seems unlikely.

```python
autos = autos[autos["registration_year"].between(1900,2016)]

print(autos['registration_year'].value_counts(normalize=True).sort_index())

print(autos['registration_year'].describe())
```

We can see after the cleaning that the median and mean are now roughly the same.

### Exploring Price by Brand

We will now aggregate the different brands to look at the price by brand. We will look at brands that have over a certain percentage of values.

To do this, we will filter on the dataframe with a map of the dataframe where the value counts of the brand are 5% or more of the total data points left.

```python
top_autos = autos[autos['brand'].map(autos['brand'].value_counts()) > len(autos)*.05]

top_autos['brand'].value_counts()
```

Now from these, we will loop over the brands and assign the mean price to the dictionary for each brand.

```python
avg_prices = {}

for val in top_autos['brand'].unique():
    auto = top_autos[top_autos['brand'] == val]
    avg_prices[val] = auto['price'].mean()
    
print(avg_prices)
```

From these, we can see that the highest average price are BMWs and the cheapest are Opels. We will also add mileage and pretty print the results.

```python
avg_mileage = {}

for val in top_autos['brand'].unique():
    auto = top_autos[top_autos['brand'] == val]
    avg_mileage[val] = auto['odometer_km'].mean()
    
avg_prices_series = pd.Series(avg_prices)
avg_mileage_series = pd.Series(avg_mileage)

df = pd.DataFrame(avg_prices_series, columns=['mean_price'])
df['mean_mileage'] = avg_mileage_series

print(df.to_string())
```

## Further Data Cleaning

Some things that we can do to clean the data set further include the following:
- Identifying categorical data that use german words, translate them, and map the values to their English counterparts
- Converting the dates to be uniform numeric data, so "2016-03-21" becomes the integer 20160321
- Seeing if there are particular keywords in the name column that you can extract as new columns

### Translating Columns

We will be using a Google Translate API to be doing this.

```python
from googletrans import Translator

translator = Translator()
```

```python
# List of columns in need of German translations
german_list = [
    'unrepaired_damage',
    'fuel_type',
    'gearbox',
    'vehicle_type',
    'offer_type',
    'seller'
]
```

```python
for col in german_list:
    # Create unique list from the column
    uniques = list(autos[col].copy().dropna().unique())
    
    # Translate these words and add to a dictionary
    translations = translator.translate(uniques, src="de")
    replace_dict = {}
    
    for word in translations:
        replace_dict[word.origin] = word.text.lower()
        
    autos[col] = autos[col].replace(replace_dict)
```

```python
autos[german_list].value_counts()
```

Here we can see that the words are now mostly translated except for anything shortened or acroynms.

### Converting Date Data

We can convert date data to uniform integers in order to be able to translate those values as needed.

```python
autos['date_crawled'] = autos['date_crawled'].str.replace('-','').astype('int32')
autos['ad_created'] = autos['ad_created'].str.replace('-','').astype('int32')
autos['last_seen'] = autos['last_seen'].str.replace('-','').astype('int32')
```

```python
autos.head()
```

### Creating New Features

We can use the last_seen and ad_created column to create the length of time that the ad has been on.

```python
autos['length_on_ad'] = autos['last_seen'] - autos['ad_created']
```

```python
autos['length_on_ad'].value_counts().sort_index()
```

Above, we can see that the shortest amount of time is 0 days and the longest amount of time was 2 months and 3 days.

### Mileage and Price Patterns

Let's split the odometer_km column into groups and aggregate to see if average prices follow patterns based on the mileage.

```python
odometer_price = autos.groupby('odometer_km')['price'].mean()
odometer_price
```

```python
odometer_price.plot(x='odometer_km', y='price', kind='scatter')
```

As you can see above, there seems to be a downard trend from 50000 km onwards. There are strange outliers in the low odometer_km values, but this might be due to strange data influencing the mean.

### Price Reduction Due to Damage

Finally, let's see how much damage does to reduce the price on average. In order to do this, we will group on brand/model and odometer_km in order to standardize those main variables as much as possible. Then we'll subtract the means of any matches.

```python
multi_autos_no_damage = autos[autos['unrepaired_damage'] == 'no'].groupby(
    ['brand_model','odometer_km'])['price'].mean().to_frame()
multi_autos_no_damage
```

```python
multi_autos_damage = autos[autos['unrepaired_damage'] == 'yes'].groupby(
    ['brand_model','odometer_km'])['price'].mean().to_frame()
multi_autos_damage
```

```python
merged_autos_damage = multi_autos_damage.merge(
    multi_autos_no_damage, left_on=['brand_model','odometer_km'], right_on=['brand_model','odometer_km'],
    suffixes=('_damage','_no_damage'))
merged_autos_damage
```

```python
merged_autos_damage['difference'] = merged_autos_damage['price_no_damage'] - merged_autos_damage['price_damage']
merged_autos_damage
```

```python
merged_autos_damage['difference'].describe()
```

Interestingly, some cars on average ended up being cheaper with the damage. However, on average, it seems that cars tend to be about $900 cheaper.
